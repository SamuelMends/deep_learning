```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from scipy.sparse import issparse
```

```python
base = pd.read_csv('autos.csv', encoding='ISO-8859-1')
base = base.drop(['dateCrawled', 'dateCreated', 'nrOfPictures', 'postalCode', 'lastSeen', 'name', 'seller', 'offerType'], axis=1)
```

```python
base = base[(base['price'] > 10) & (base['price'] < 350000)]
```

```python
valores = {'vehicleType': 'limousine', 'gearbox': 'manuell', 'model': 'golf', 'fuelType': 'benzin', 'notRepairedDamage': 'nein'}
base = base.fillna(value=valores)
```

```python
previsores = base.iloc[:, 1:].values  # exclui a coluna 'price'
preco_real = base.iloc[:, 0].values   # coluna 'price'
```

```python
categorical_columns = [0, 1, 3, 5, 8, 9, 10]
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(), categorical_columns)],
    remainder='passthrough')
previsores = ct.fit_transform(previsores)
```

```python
previsores = ct.fit_transform(previsores)
if issparse(previsores):
    previsores = previsores.toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(previsores, preco_real, test_size=0.2, random_state=0)
```

```python
tensor_X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
tensor_y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
tensor_X_test = tf.convert_to_tensor(X_test, dtype=tf.float32)
tensor_y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)
```
